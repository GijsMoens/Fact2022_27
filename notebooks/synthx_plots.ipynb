```python
#IMPORTS
import numpy as np
import matplotlib.pyplot as plt
```

but exclude some of the moonwalk runs

synthx 3 (seed 42) groups (300, 125, 75)

```python
#EXTRACTING DATA
PATH = '../data/synthx3_42/300,125,75/infmaxresults/'
METHODS = ['deepwalk', 'fairwalk', 'crosswalk', 'moenswalk_1', 'moenswalk_2', 'moenswalk_3', 'moenswalk_4']
ROOTNAME = 'synth500_Phom0.01-0.05_Phet0.001-0.005'
ITERS = [1,2,3,4,5]
NUM_NODES = 500
NUM_GROUPS = 3
NODES_PER_GROUP = [300, 125, 75]

def read_file_to_list(fn):
    file = open(fn, 'r')
    file_as_list = file.read().split(' ')
    return file_as_list

def extract_data(fal, num_groups=NUM_GROUPS):
    data = {}
    data['total'] = float(fal[0])
    
    for n in range(num_groups):
        data[str(n)] = float(fal[2 * (n + 1) ])
    
    return data

results = {method:{} for method in METHODS}

for method in METHODS:
    for iter in ITERS:
        filename = PATH + method + '/' + ROOTNAME + '_' + method + '_' + str(iter) + '_results.txt'
        fal = read_file_to_list(filename)
        data = extract_data(fal)
        results[method][str(iter)] = data

#COMPUTATIONS
means_total = {method: np.array([results[method][str(iter)]['total'] for iter in ITERS]).mean() / NUM_NODES * 100 for method in METHODS}
disparities = {method:np.var([np.array([results[method][str(iter)][str(n)] for iter in ITERS]).mean() / NODES_PER_GROUP[n] * 100 
                    for n in range(NUM_GROUPS)]) for method in METHODS}

#PLOTTING
labels = ['deepwalk', 'fairwalk', 'crosswalk', 'moonwalk_1', 'moonwalk_2', 'moonwalk_3', 'moonwalk_4'] 
infs = means_total.values()
disps = disparities.values()

x = np.arange(len(labels))
width = 0.25
gap = 0.02

infs_color = 'green'
disps_color = 'red'

fig, ax = plt.subplots()

ax2 = ax.twinx()
rects1 = ax.bar(x - width/2 - gap/2, infs, width, label='total influence (%)', color=infs_color)
rects2 = ax2.bar(x + width/2 + gap/2, disps, width, label='disparity', color=disps_color)

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)

ax.set_ylabel('total influence (%)', c=infs_color)
ax2.set_ylabel('disparity', c=disps_color)
ax.tick_params(axis='y', colors=infs_color)
ax2.tick_params(axis='y', colors=disps_color)

plt.tight_layout()
plt.savefig('../figures/synthx3_42_500_infmax.pdf')



        
```
